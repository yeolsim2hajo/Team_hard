```python
# codeFestival_1. 97ë²ˆ

n = int(input())
ln = [0]*n
lst = list(map(int, input().split()))
lst.sort(reverse = True)
for i in range(n):
    ln[i] = lst[i]
rst = 0
def solution(a):
    global rst
    while a > -n:
        for i in range(n):
            ln[i] -= 1
            if ln[i] == 0:
                a -= 1
                if a < 0:
                    continue
                ln[i] = lst[len(lst)-a-1]
        rst += 1
solution(len(lst)-n)
print(rst)
```
