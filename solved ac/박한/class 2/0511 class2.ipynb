```python
# 백준 class 2++ 1929 소수 구하기

m, n = map(int, input().split())
def abc(a):
    if a > 3:
        rst = 0
        for i in range(2, a//2+1):
            if a%i == 0:
                rst += 1
                break
        if not rst:
            print(a)
    elif a > 1:
        print(a)
for k in range(m, n+1):
    abc(k)
```
